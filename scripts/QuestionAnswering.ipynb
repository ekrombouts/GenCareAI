## Setup

### Laad libraries

```python
# Duurt ook nog best een tijdje, 20 sec ofzo
import pandas as pd
from transformers import pipeline 
```

```python
vb_rap = 6
```

### Laad data

```python
df = pd.read_csv("../zorgdata/gen_data.csv")
```

### Question answering Henry K Bert base multilingual

```python
vb_rapportage = df['rapportage'][vb_rap]
print(vb_rapportage)
```

```python
model="henryk/bert-base-multilingual-cased-finetuned-dutch-squad2"

qa_pipeline = pipeline(
    "question-answering",
    model = model,
    tokenizer = model
)

qa_pipeline({
    'context': vb_rapportage,
    'question': "Welke ADL hulp heeft de cliënt nodig?"})
```

```python
# Functie die op elke rij zal worden toegepast
def answer_question(row):
    result = qa_pipeline({
        'context': row['rapportage'],
        'question': "Welke ADL hulp heeft deze client nodig?"
    })
    return result['answer']
```

```python
# Pas toe op de dataset
df['antwoord'] = df.apply(answer_question, axis=1)
```
