```python
import pandas as pd
from datasets import load_dataset

import os
```

```python
ds_clients = load_dataset("ekrombouts/Galaxy_clients")
ds_records = ds = load_dataset("ekrombouts/Galaxy_records")
```

```python
df_clients = ds_clients['train'].to_pandas() 
df_records = ds_records['train'].to_pandas()
```

```python
df_cr = pd.merge(df_records, df_clients, on='ct_id', how='left') 
df_cr = df_cr[['ct_id', 'name', 'datetime', 'note']]
```
